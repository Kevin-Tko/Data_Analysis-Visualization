```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "coffee/total-production.csv"

coffee_production = pd.read_csv(path)
```

# Coffee Production Stats

* Create a bar chart of the top 10 coffee producers by total production in 'total-production.csv', sorted from highest to lowest. Change the xticks to two letter country abbreviations.


* Data Prep Code

<mark style="background-color: #000000"> coffee_production = pd.read_csv("coffee/total-production.csv").T </mark> 

<mark style="background-color: #000000"> coffee_production.columns = coffee_production.iloc[0] </mark> 

<mark style="background-color: #000000"> coffee_production = coffee_production.drop("total_production") </mark> 


```python
coffee_production.head()
```

```python
coffee_production = coffee_production.T #Transposing the dataframe so that years will be my index values
coffee_production.head()
```

```python
coffee_production.columns=coffee_production.iloc[0] #Setting first row of my DF as the column names
coffee_production_df = coffee_production[1:] #dropping the first row from the DF after setting it as the col names
# coffee_production.drop(labels='total_production', axis=0, inplace=True) # equal to step 2
coffee_production_df.head()
```

```python
coffee_production_df.info()
```

```python
top10_producers = coffee_production_df.sum().sort_values(ascending=False).iloc[:10]

top10_producers
```

```python
fig,ax=plt.subplots()

ax.bar(x=top10_producers.index,
       height=top10_producers /1000000
      )

ax.xaxis.set_tick_params(rotation=30) # Rotate X axis labels
ax.spines[['top','right']].set_visible(False) #remove top and right borders
ax.set_ylabel('Production(60KG bags in Millions)',
               fontfamily='Garamond'
              )
ax.set_title('To 10 Coffee Producers',
            fontfamily='Garamond'
           );
```

# Top 5 over time

* Create a line chart with the top 5 nations over time. Make sure to include a legend! Consider converting the index to datetime64[ns].

```python
top5 = coffee_production_df.loc[:, ['Brazil','Viet Nam','Colombia','Indonesia','Ethiopia']]

top5.index = top5.index.astype('datetime64[ns]')

top5.head()
```

```python
fig, ax = plt.subplots()


ax.set_ylim(0,70)
ax.plot(top5.index, top5['Brazil'] / 1000, label='Brazil')
ax.plot(top5.index, top5['Viet Nam'] / 1000, label='Viet Nam')
ax.plot(top5.index, top5['Colombia'] / 1000, label='Colombia')
ax.plot(top5.index, top5['Indonesia'] / 1000, label='Indonesia')
ax.plot(top5.index, top5['Ethiopia'] / 1000, label='Ethiopia')

ax.legend(frameon=False)
ax.spines[['top','right']].set_visible(False)
ax.set_ylabel('Production (Thousands of Bags)',
              fontfamily='Garamond'
             )
fig.suptitle('Top 5 coffee producing nations 1990-2018',
             fontfamily='Garamond',
             fontsize=16
            )
ax.set_title('Brazil takes the lead as Viet Nam surges to 2nd spot',
             fontfamily='Garamond'
            );
```

# COMPOSITION OVER TIME

Our client is interested in looking at how Brazil's share of global coffee production has changed over time.

Build a plot that global total production as well as Brazil's contribution to global production over time.

```python
brazil_vs_others = (coffee_production_df
                    .assign(others = coffee_production_df
                                               .drop('Brazil',axis=1).sum(axis=1))
                   ).astype({'Brazil':'float64', 'others':'float64'})[['Brazil','others']]
brazil_vs_others.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))

ax.stackplot(brazil_vs_others.index.astype('datetime64[ns]'),
             brazil_vs_others['Brazil'],
             brazil_vs_others['others'],
             labels=['Brazil', 'Others']
            )

ax.spines[['top', 'right']].set_visible(False)
ax.xaxis.set_tick_params(rotation=40)
ax.legend(bbox_to_anchor=(0.1,1));
fig.suptitle("Brazil's Share of Global Coffee Production 1990-2018",
             fontfamily='Garamond',
              fontsize=18
            )
ax.set_title('Brazil Increases Share of Growing Market',
             fontfamily='Garamond',
             fontsize=16
            )
ax.set_ylabel('Production(60 KG bag)',
              fontfamily='Garamond',
              fontsize=16
             );
```

## PRODUCTION RELATIONSHIPS

We want to explore how Brazil compares to its competitors in terms of production. Build a scatterplot  comparing Brazil Vs. VietNam Production and one comparing Brazil to Venezuela.

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.scatter(coffee_production_df['Brazil'],
           coffee_production_df['Viet Nam'],
           alpha=0.6
          )
ax.set_title("Brazil and Vietnam Production Continue to Increase",
             fontfamily='Garamond',
             fontsize=18
            );

ax.set_ylabel('VietNam Production(Millions 60KG Bags)',
              fontfamily='Garamond',
              fontsize=14
             )
ax.set_xlabel('Brazil Production(Millions 60KG Bags)',
              fontfamily='Garamond',
              fontsize=14
             );
ax.spines[['top', 'right']].set_visible(False)
```

```python
fig, ax = plt.subplots()

ax.scatter(coffee_production_df['Brazil'],
           coffee_production_df['Venezuela'],
           alpha=0.6
          )
ax.set_title("Venezuela Production Drops as Brazil Production Continue to Increase",
             fontfamily='Garamond',
             fontsize=18
            );
ax.set_ylim(0, 1800)
ax.set_xlim(0, 70000)

ax.set_ylabel('Venezuela Production(Millions 60KG Bags)',
              fontfamily='Garamond',
              fontsize=14
             )
ax.set_xlabel('Brazil Production(Millions 60KG Bags)',
              fontfamily='Garamond',
              fontsize=14
             );
ax.spines[['top', 'right']].set_visible(False)
```

# COMPOSITION OF COFFEE PRODUCTION

* Create a Pie chart with the top 5 producers, as well as the sum for the rest of countries.

* Finally, create two donut charts, one representing Brazil's share of global production in 1990, and one for Brazil's share in 2018.

```python
Other_producers=coffee_production_df.drop(['Brazil', 'Viet Nam', 'Colombia', 'Indonesia', 'Ethiopia'], axis=1).sum(axis=1)

top5['other_producers']=Other_producers.values

coffee_production_top5_2018 = top5.loc['2018-01-01']

coffee_production_top5_2018
```

```python
fig, ax = plt.subplots()

ax.pie(coffee_production_top5_2018,
       labels=['Brazil', 'Viet Nam', 'Colombia', 'Indonesia', 'Ethiopia','other_producers'],
       startangle=90,
       autopct="%.0f%%",
       explode=(0.1,0,0,0,0,0)
      );
ax.set_title('Share of Global Coffee Production in 2018',
             fontfamily='Garamond',
             fontsize=16
            );
```

```python
brazil_vs_others.head()
```

```python
brazil_vs_others_1990 = brazil_vs_others.loc['1990']
brazil_vs_others_1990
```

```python
fig, ax = plt.subplots()

ax.pie(brazil_vs_others_1990,
       labels=['Brazil','others'],
       startangle=90,
       autopct="%.0f%%",
       pctdistance=.85
      )
donut_hole=plt.Circle((0,0), 0.70, fc='white')
fig=plt.gcf()
fig.gca().add_artist(donut_hole);

ax.set_title('Brazil Share of Global Production in 1990',
             fontsize=16,
             fontfamily='Garamond'
            );
```

```python
fig, ax = plt.subplots()

ax.pie(brazil_vs_others_1990,
       labels=['',''],
       startangle=90,
       pctdistance=.85,
       colors=['blue', 'white']
      )
donut_hole=plt.Circle((0,0), 0.70, fc='white')
fig=plt.gcf()
fig.gca().add_artist(donut_hole)

ax.set_title('Brazil Ratio to Total Global Production in 1990',
             fontsize=16,
             fontfamily='Garamond'
            )

ratio = round(brazil_vs_others_1990.loc['Brazil']/ (brazil_vs_others_2018.loc[['Brazil','others']].sum())*100)

plt.text(0,
         0,
         f'{ratio}%',
         ha='center',
         va='center',
         fontsize=42
);
```
